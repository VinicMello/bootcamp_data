```python
import os
import json
import numpy as np
import pandas as pd
```

```python
# Diretório atual e caminho para os dados
dir = os.getcwd()
data = os.path.join(dir, 'data')
```

```python
# Lê o arquivo JSON de dados de hospedagem
with open(os.path.join(data, 'dados_hospedagem.json'), 'r') as f:
    dados_json = json.loads(f.read())

# Converte a parte 'info_moveis' do JSON em um DataFrame
df = pd.json_normalize(dados_json, record_path='info_moveis')
df.head(1)
```

```python
columns_df = list(df.columns)
print(columns_df)

# Explode colunas a partir da 4ª posição (transforma listas em linhas)
df = df.explode(columns_df[3:])
df.head(1)
```

```python
# Reseta os índices do DataFrame, removendo o índice antigo
df.reset_index(inplace=True, drop=True)
df.head(1)
```

```python
# Exibe informações sobre o DataFrame
df.info()
```

```python
# Define colunas numéricas que devem ser convertidas para inteiros
col_numeric = ['max_hospedes', 'quantidade_banheiros','quantidade_quartos','quantidade_camas']

# Converte colunas para tipos numéricos apropriados (int64 e float64)
df[col_numeric] = df[col_numeric].astype(np.int64)
df['avaliacao_geral'] = df['avaliacao_geral'].astype(np.float64)

# Exibe informações sobre o DataFrame
df.info()
```

```python
# Seleciona e exibe as colunas relacionadas a valores monetários($)
columns = ['taxa_deposito', 'taxa_limpeza', 'preco']
df[columns]
```

```python
# Remove símbolos e formata valores monetários para float64
df[columns] = df[columns].applymap(lambda row: row.replace('$', '').replace('.', '').replace(',', '.').strip()).astype(np.float64)
df[columns]
```

```python
# Exibe informações sobre o DataFrame
df.info()
```

```python
# Converte texto da coluna 'descricao_local' para letras minúsculas
df['descricao_local'] = df['descricao_local'].str.lower()
df['descricao_local']
```

```python
# Remove caracteres que não sejam letras, números, hífen ou apóstrofo, substituindo por espaço
# a-z: todas as letras, do "a" ao "z", minúsculas;
# A-Z: todas as letras, do "A" ao "Z", maiúsculas;
# 0-9: todos os números, de 0 a 9;
# \-: hífen;
# \': apóstrofo.
df['descricao_local'] = df['descricao_local'].str.replace('[^a-zA-Z0-9\-\']', ' ', regex=True)
df['descricao_local']
```

```python
# Remove hífens isolados que não estão entre caracteres alfanuméricos
# (?<!\w): uma "lookbehind" negativa que verifica se há um caractere de palavra antes do hífen ou se há apenas um espaço em branco, resultando em "verdadeiro" se não houver nada;
# (?!\w): uma lookahead negativa que verifica se há um caractere de palavra depois do hífen.
df['descricao_local'] = df['descricao_local'].str.replace('(?<!\w)-(?!\w)', '', regex=True)
df['descricao_local'] 
```

```python
# Separa texto em listas de palavras, removendo espaços extras
# Tokenização: separa as palavras em uma lista, removendo espaços extras.
df['descricao_local'] = df['descricao_local'].str.split()
df.head(2)
```

```python
# Remove chaves e aspas da coluna 'comodidades'
df['comodidades'] = df['comodidades'].str.replace(r'\{|}|\"', '', regex=True)
df['comodidades']
```

```python
# Separa a string da coluna 'comodidades' em listas usando vírgula como delimitador
df['comodidades'] = df['comodidades'].str.split(',')
df.head(1)
```

```python
# Converte texto da coluna 'descricao_vizinhanca' para minúsculas
df['descricao_vizinhanca'] = df['descricao_vizinhanca'].str.lower()

# Remove caracteres especiais e hífens isolados da coluna
df['descricao_vizinhanca'] = df['descricao_vizinhanca'].str.replace('[^a-zA-Z0-9\-\']', ' ', regex=True)
df['descricao_vizinhanca'] = df['descricao_vizinhanca'].str.replace('(?<!\w)-(?!\w)', '', regex=True)

# Tokeniza o texto, transformando em lista de palavras
df['descricao_vizinhanca'] = df['descricao_vizinhanca'].str.split()
df.head(1)
```

```python
# Dados de móveis disponíveis
df_moveis_disponiveis = pd.read_json(os.path.join(data, 'dados_moveis_disponiveis.json'))
df_moveis_disponiveis.head(5)
```

```python
# Exibe informações sobre o DataFrame
df_moveis_disponiveis.info()
```

```python
# Preenche valores nulos na coluna 'preco' com '0.0'
df_moveis_disponiveis['preco'] = df_moveis_disponiveis['preco'].fillna('0.0')

# Remove '$' e vírgulas da coluna 'preco'
df_moveis_disponiveis['preco'] = df_moveis_disponiveis['preco'].apply(lambda x: x.replace('$', '').replace(',', ''))

# Converte a coluna 'preco' para tipo float64
df_moveis_disponiveis['preco'] = df_moveis_disponiveis['preco'].astype(np.float64)

df_moveis_disponiveis.head()
```

```python
# Converte a coluna 'data' para formato datetime
df_moveis_disponiveis['data'] = pd.to_datetime(df_moveis_disponiveis['data'])
df_moveis_disponiveis.info()
```

```python
# Subset - Agrupa dados por mês (YYYY-MM) e soma a coluna 'vaga_disponivel'
subset = df_moveis_disponiveis.groupby(df_moveis_disponiveis['data'].dt.strftime('%Y-%m'))['vaga_disponivel'].sum()
subset.head()
```
