```python
# Imports
import os
import json
import numpy as np
import pandas as pd
```

```python
# Environment Variables
dir = os.getcwd()
data = os.path.join(dir, 'data')
```

```python
#
with open(os.path.join(data, 'dados_hospedagem.json'), 'r') as f:
    dados_json = json.loads(f.read())

#
df = pd.json_normalize(dados_json, record_path='info_moveis')
df.head(1)
```

```python
#
columns_df = list(df.columns)
print(columns_df)

#
df = df.explode(columns_df[3:])
df.head(2)
```

```python
#
df.reset_index(inplace=True, drop=True)
df.head(2)
```

```python
#
df.info()
```

```python
#
col_numeric = ['max_hospedes', 'quantidade_banheiros','quantidade_quartos','quantidade_camas']

#
df[col_numeric] = df[col_numeric].astype(np.int64)
df['avaliacao_geral'] = df['avaliacao_geral'].astype(np.float64)
df.info()
```

```python
#
columns = ['taxa_deposito', 'taxa_limpeza', 'preco']
df[columns]
```

```python
#
df[columns] = df[columns].applymap(lambda row: row.replace('$', '').replace('.', '').replace(',', '.').strip()).astype(np.float64)
df[columns]
```
